As we can see that columns ***ID, URL, Name, Icon URL and Subtitle*** having all distinct values and do not effect the rating of the apploication. So we can drop them.

```python
data = data.drop("URL", axis=1)
data = data.drop("Name", axis=1)
data = data.drop("Subtitle", axis=1)
data = data.drop("Icon URL", axis=1)
data = data.drop("ID", axis=1)
```

```python
data.describe(include="all")
```

Also we don't need ***Description, Developer, Age Rating and any Release Date*** to predict the rating of any product.

```python
data=data.drop("Developer", axis=1)
data=data.drop("Original Release Date", axis=1)
data=data.drop("Current Version Release Date", axis=1)
data=data.drop("Description",axis=1)
data=data.drop("Age Rating", axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
genre_counts = dict(data['Primary Genre'].value_counts())
```

```python
distributed_genres={}
for i in genre_counts.keys():
    distributed_genres[i]=data[data['Primary Genre']==i]
```

```python
distributed_genres.keys()
```

```python
sns.distplot(distributed_genres['Games']['Average User Rating'], kde=False)
```

```python
for i in distributed_genres.keys():
    sns.distplot(distributed_genres[i]['Average User Rating'], kde=False)
    plt.title(i)
    plt.show()
```

#### In the above Histograms we can see that the value count of Games genre is the highest in Primary Genre feature and it has most of the ratings between 4.0 and 5.0 so ***Games*** is the most significant genre in this dataset.
