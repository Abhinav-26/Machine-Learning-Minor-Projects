```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies_csv = "/Users/abhinavdubey/Trainings/IBM Courses/ Cognitive AI Codes/movies.csv"
ratings_csv = "/Users/abhinavdubey/Trainings/IBM Courses/ Cognitive AI Codes/ratings.csv"
```

```python
movies_dataset = pd.read_csv(movies_csv)
ratings_dataset = pd.read_csv(ratings_csv)
```

```python
ratings_dataset
```

```python
#Now we need to extract years from the title column such that it doesnot hamper the year written on the movie name.
movies_dataset['year'] = movies_dataset.title.str.extract('(\(\d\d\d\d\))',expand=False)
movies_dataset
```

```python
#In the dataset we can observer that we have () also in the year column, so we need to remove it.
movies_dataset['year'] = movies_dataset.title.str.extract('(\d\d\d\d)',expand=False)
movies_dataset
```

```python
#Now we need to remove the years and replace it using strip method to avoid any WhiteSpaces

#Replacing WhiteSpaces with empty string ''
movies_dataset['title'] = movies_dataset.title.str.replace('(\(\d\d\d\d\))', '')

#Applying Strip method to get rid of any white spaces 
movies_dataset['title'] = movies_dataset['title'].apply(lambda x: x.strip())

movies_dataset
```
