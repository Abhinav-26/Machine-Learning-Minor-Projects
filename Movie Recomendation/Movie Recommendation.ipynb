```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies_csv = "/movies.csv"
ratings_csv = "/ratings.csv"
```

```python
movies_dataset = pd.read_csv(movies_csv)
ratings_dataset = pd.read_csv(ratings_csv)
```

```python
ratings_dataset
```

```python
#Now we need to extract years from the title column such that it doesnot hamper the year written on the movie name.
movies_dataset['year'] = movies_dataset.title.str.extract('(\(\d\d\d\d\))',expand=False)
movies_dataset
```

```python
#In the dataset we can observer that we have () also in the year column, so we need to remove it.
movies_dataset['year'] = movies_dataset.title.str.extract('(\d\d\d\d)',expand=False)
movies_dataset
```

```python
#Now we need to remove the years and replace it using strip method to avoid any WhiteSpaces

#Replacing WhiteSpaces with empty string ''
movies_dataset['title'] = movies_dataset.title.str.replace('(\(\d\d\d\d\))', '')

#Applying Strip method to get rid of any white spaces 
movies_dataset['title'] = movies_dataset['title'].apply(lambda x: x.strip())

movies_dataset
```

```python
#In our genres column we can see we have genres which is not organised so first we will split these genres into 
#diffrent parts using split method and then we will use One Hot encoding techniques to more simplify the columns where
#each column corresponds to one genre.

#Splitting the genre column
movies_dataset['genres'] = movies_dataset.genres.str.split('|')
movies_dataset
```

```python
#Applying One Hot Encoding

#Since we need the Movies_dataset at first so we will not perform One Hot encoding in it
moviesWithGenres_df = movies_dataset.copy()

#For every row in the dataframe, iterate through the list of genres and place a 1 into the corresponding column
for index, row in movies_dataset.iterrows():
    for genre in row['genres']:
        moviesWithGenres_df.at[index, genre] = 1
        
#Filling in the NaN values with 0 to show that a movie doesn't have that column's genre
moviesWithGenres_df = moviesWithGenres_df.fillna(0)
moviesWithGenres_df
```

```python
#Now lets see about ratings dataset.
ratings_dataset
```

```python
ratings_dataset=ratings_dataset.drop("timestamp", axis=1)
ratings_dataset
```

## Recommended System 

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
#Filtering the movies from Moviedataset that we have in InputId
inputId = movies_dataset[movies_dataset['title'].isin(inputMovies['title'].tolist())]

inputId
```

```python
inputMovies = pd.merge(inputId, inputMovies)

inputMovies
```

```python
inputMovies = inputMovies.drop('genres', 1).drop('year', 1)

inputMovies
```

```python
userMovies = moviesWithGenres_df[moviesWithGenres_df['movieId'].isin(inputMovies['movieId'].tolist())]

userMovies
```

```python
userMovies = userMovies.reset_index(drop=True)

userMovies
```

```python
userGenreTable = userMovies.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)

userGenreTable
```

```python
inputMovies['rating']
```

```python
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])

userProfile
```

```python
genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['movieId'])

genreTable
```

```python
genreTable = genreTable.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)

genreTable
```

```python
genreTable.shape
```

```python
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())

recommendationTable_df
```

```python
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)

recommendationTable_df
```
