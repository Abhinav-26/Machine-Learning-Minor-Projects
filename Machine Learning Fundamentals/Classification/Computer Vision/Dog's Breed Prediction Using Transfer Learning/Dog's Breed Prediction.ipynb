# Transfer Learning

I will be using the pre trained models for making my predition and I will be using three models ie, VGG16, Inception_v3, ResNet50. For this project I will be using imagnet as my weights.

# VGG16
In this code blocks I have used the pre-trained model of VGG16 for making the dog's breed prediction easy. VGG16 is a very powerful model with total of 16 deep layers of Network. Transfer Learning is uses the concept of implementing the pre-trained models for our own usages.

```python
from keras.applications import vgg16
```

```python
#### We are using imagenet as weights, it will download if u don't have by default.
model = vgg16.VGG16(weights='imagenet')
```

```python
 model.layers[0].input # To check the input Shape used in this model
```

```python
from keras.preprocessing import image
img = image.load_img("dog.jpg", target_size=(224,224))
```

```python
img
```

```python
type(img) 
```

Since we are using a model which is already being created by someone, so we need specify the size of image that we are providing as per the size which was given while training the model. Now in CNN we cannot use Pillow datatype of an image for different manipulations, so we need to convert it into numpy array, as in down code blocks we did. Also CNN accepts 4 dimentional image so we need to convert it into 4D, as we did. In the down code blocks, decode_predictions will convert the predicted value into text value so that we can easily understand it.

```python
img_np = image.img_to_array(img) #Convertion of image into numpy array
```

```python
print(img_np.shape)
type(img_np)
```

```python
import numpy as np
```

```python
img_converted = np.expand_dims(img_np, axis=0)
img_converted.shape
```

```python
from keras.applications.vgg16 import decode_predictions
pred = model.predict(img_converted)
pred
```

```python
decode_predictions(pred, top=3)[0]
```

### The Concept Used above for using VGG16 will be the same for all other weights, just we need to import the model and load import decode_predictions and preprocessing_input from that model.

# Inception_V3
Below in these Code Blocks I will be using the Inception_V3 pre-trained model for making dogs breed prediction. Inception_v3 is being created by Google which is also one of the most powerful Deep Learning Convolutional Architecture.

```python
from keras.applications import inception_v3
```

```python
InceptionModel = inception_v3.InceptionV3(weights="imagenet")
```

```python
InceptionModel.layers[0].input
```

```python
img1 = image.load_img("Pug.jpeg", target_size=(299,299))
img1
```

```python
print(type(img1))
imgTo_npConversion = image.img_to_array(img1)
print(type(imgTo_npConversion))
```

```python
img4D = np.expand_dims(imgTo_npConversion, axis=0)
print(img4D.shape)
```

```python
from keras.applications.inception_v3 import decode_predictions
from keras.applications.inception_v3 import preprocess_input
finalImg = preprocess_input(img4D)
predIncep = InceptionModel.predict(finalImg)
decode_predictions(predIncep)
```

# ResNet50
In the below code blocks we will be using ResNet50 Model for making prediction of our dog's breed. ResNet50 is also one of the most commently used and powerful pre-trained models which we can use to ease our tasks. 

```python
from keras.applications import resnet50
```

```python
ModelResNet = resnet50.ResNet50(weights="imagenet")
```

```python
ModelResNet.layers[0].input
```

```python
img2 = image.load_img("German Shepherd.jpg", target_size=(224,224))
img2
```

```python
print(type(img2))
imgIntoArray = image.img_to_array(img2)
print(type(imgIntoArray))
```

```python
CNNImage = np.expand_dims(imgIntoArray, axis=0)
type(CNNImage)
```

```python
from keras.applications.resnet50 import decode_predictions
from keras.applications.resnet50 import preprocess_input
fImgage = preprocess_input(CNNImage)
ResNetPred = ModelResNet.predict(fImgage)
decode_predictions(ResNetPred)
```
