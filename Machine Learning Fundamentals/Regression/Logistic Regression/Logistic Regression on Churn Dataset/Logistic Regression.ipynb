## Feature Selection & Feature Engineering
Implementing Univariate Feature Selection Method for selection best features among all the 28 columns that we have in our dataset. Using heat and trail method for selecting the best no. of features for getting better accuarcy and storing those best features in variable named X_SelectedFeatures which we will be using further for tarining our model.
After selecting the best features we will be using some feature engineering concepts like Standardization from standard scaler and then making our dataset ready for tarining and then for prediction. 

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
dataset.columns
```

```python
dataset = dataset.values
```

```python
X = dataset[:,0:26]
y = dataset[:,27]
```

```python
test = SelectKBest(score_func=chi2, k=6)        # k is number of features
fit = test.fit(X, y)
```

```python
np.set_printoptions(precision=3)
print(fit.scores_)
```

```python
X_SelectedFeatures = fit.transform(X)
print(X_SelectedFeatures)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_SelectedFeatures, y, test_size=0.33, random_state=42)
```

```python
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
X_train_scaled
```

## Model Training & Prediction
Training the model by using the concepts of Logistic Regression and then making prediction from the data which we splitted for testing. Also calculating the accuracy socre of the prediction.

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.metrics import jaccard_similarity_score
jaccard_similarity_score(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the model is :",accuracy_score(y_test, y_pred))
```
