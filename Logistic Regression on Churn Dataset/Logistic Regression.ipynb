```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
dataset.columns
```

```python
dataset = dataset.values
```

```python
X = dataset[:,0:26]
y = dataset[:,27]
```

```python
test = SelectKBest(score_func=chi2, k=10)        # k is number of features
fit = test.fit(X, y)
```

```python
np.set_printoptions(precision=3)
print(fit.scores_)
```

```python
X_SelectedFeatures = fit.transform(X)
print(X_SelectedFeatures)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_SelectedFeatures, y, test_size=0.33, random_state=42)
```

```python
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
X_train_scaled
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

```python
y_pred = LR.predict(X_test)
```

```python
from sklearn.metrics import jaccard_similarity_score
jaccard_similarity_score(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the model is :",accuracy_score(y_test, y_pred))
```
